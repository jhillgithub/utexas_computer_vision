```python
# import all the things
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
print("numpy version", np.__version__)
print("opencv version", cv2.__version__)
```

```python
# Read in an image using opencv (cv2) imread method and plot it with matplotlib
img = cv2.imread('../../Images/beefsteak.jpg', cv2.IMREAD_COLOR)
plt.imshow(img)
plt.show()
#???
# What happened here?
```

```python
# OpenCV uses BGR pixel ordering while Matplotlib expects RGB ordering
# Here we can uses the cvtColor method to convert the image into RGB ordering for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imshow(img_rgb)
```

```python
# That's handy. Let's save that code in a helper method to use again
def show(img):
    """convert opencv BGR ordering to RGB ordering"""
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imshow(img_rgb)
```

```python
show(img)
```

```python
# What does an image look like from python's perspective?
print("Image type: ", type(img))
print("Image data type: ", img.dtype)
print("Image shape: ", img.shape)
```

```python
# Let's make a copy of the image data and make some changes
tmp = img.copy()
blues = tmp[:,:,0]
greens = tmp[:,:,1]
reds = tmp[:,:,2]

# Let's scale back the reds to a darker color
tmp[:,:,2] = reds*0.5
print("Reds max value: ", reds.max())
show(tmp)
```

```python
# What if we wanted to change individual pixels?
tmp[100,100] = [255, 255, 255]
show(tmp)
```

```python
# Looks good. Let's save it.
cv2.imwrite('new_beefsteak.jpg', tmp)
```
