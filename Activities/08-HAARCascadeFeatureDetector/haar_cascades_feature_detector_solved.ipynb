```python
# import all the things
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
print("numpy version", np.__version__)
print("opencv version", cv2.__version__)
```

```python
def show(img):
    """plot the image scaled to 15 x 15"""
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15, 15))
    plt.imshow(img_rgb, interpolation='nearest')
```

```python
# Read in image and plot it
nic = cv2.imread('../../Images/nic.jpg',cv2.IMREAD_COLOR)
show(nic)
```

```python
# Initialize face cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascade
```

```python
# Detect faces using face_cascade. Adjust settings as needed.
faces = face_cascade.detectMultiScale(
    nic,
    scaleFactor=1.3,
    minNeighbors=5,
    minSize=(10,10)
)
faces
```

```python
# Loop through all detected faces and draw bounding box for each face
for(x, y, w, h) in faces:
    cv2.rectangle(nic, (x,y), (x+w, y+h), (255,255,0), 2)
show(nic)
```

```python
# Choose a new feature to detect using a different HAAR Cascade file from
# https://github.com/opencv/opencv/tree/master/data/haarcascades
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
eye_cascade
```

```python
# Detect all eyes in the image
eyes = eye_cascade.detectMultiScale(
    nic,
    scaleFactor=1.3,
    minNeighbors=5
)
eyes
```

```python
# For each eye detected, draw a bounding box around the eye
for(x, y, w, h) in eyes:
    cv2.rectangle(nic, (x,y), (x+w, y+h), (0,255,0), 2)
show(nic)
```

```python
# Bonus: Choose image with multiple faces to test the detector
starwars = cv2.imread('../../Images/nic_star_wars.jpg',cv2.IMREAD_COLOR)
show(starwars)
```

```python
# Detect faces using face_cascade. Adjust settings as needed.
faces = face_cascade.detectMultiScale(
    starwars,
    scaleFactor=1.3,
    minNeighbors=4,
    minSize=(10,10)
)
faces
```

```python
for(x, y, w, h) in faces:
    cv2.rectangle(starwars, (x,y), (x+w, y+h), (255,255,0), 2)
show(starwars)
```
