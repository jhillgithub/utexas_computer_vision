```python
# import all the things
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import imshow
print("numpy version", np.__version__)
print("opencv version", cv2.__version__)
```

```python
def show(img):
    """convert opencv BGR ordering to RGB ordering"""
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    imshow(img_rgb)
```

```python
# load an image of a car
img = cv2.imread('../../Images/car.jpg', cv2.IMREAD_COLOR)
show(img)
```

```python
# Remember how we can select a specific pixel using bracket selection?
img[100,200]
```

```python
# Well, it turns out that you can also select a range of pixels or region of interest (ROI)
plate = img[100:150, 300:400]
show(plate)
```

```python
# Let's grab the ROI for a wheel
wheel = img[120:200,180:240]
show(wheel)
```

```python
# You can copy and paste the ROI as long as the shape matches
print("Dimensions of a wheel ROI: ", wheel.shape)
```

```python
# Use the wheel shape to select a region to paste the wheel
tmp = img.copy()
tmp[0:80, 0:60] = wheel
show(tmp)
```

```python
# Try it again with a different set of coordinates
tmp = img.copy()
tmp[100:180, 400:460] = wheel
show(tmp)
```
